```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
votes = [345,123,34,56,9]
parties = ['BJP','CONG','AAP','OTHERS','LEFT']
```

```python
plt.pie(votes, labels = parties)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(votes, labels = parties,startangle=90,counterclock=False,
       autopct='%1.2f%%')
plt.legend()
plt.show()
```
