```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
votes = [345,123,34,56,9]
parties = ['BJP','CONG','AAP','OTHERS','LEFT']
```

```python
plt.pie(votes, labels = parties)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(votes, labels = parties,startangle=90,counterclock=False,
       autopct='%1.2f%%')
plt.legend()
plt.show()
```

```python
df = pd.read_json('https://raw.githubusercontent.com/brainmentorspvtltd/NITTTR_DataAnalysisOnlineFDP/master/states_daily.json')
```

```python
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
plt.bar(df['status'], df['dl'])
plt.show()
```

```python
plt.barh(df['status'], df['dl'])
plt.show()
```

```python
confirmed_cases = df[df['status'] == 'Confirmed']
```

```python
plt.figure(figsize=(10,6))
plt.plot(confirmed_cases['date'], confirmed_cases['dl'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(confirmed_cases['date'], confirmed_cases['dl'], label='Delhi',linestyle='--')
plt.plot(confirmed_cases['date'], confirmed_cases['mh'], label='Maharsatra',linestyle='--')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(confirmed_cases['date'], confirmed_cases['mh'], label='Maharstra')
plt.bar(confirmed_cases['date'], confirmed_cases['dl'], label='Delhi')
plt.legend()
# plt.xticks(rotation='vertical')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
figure = plt.figure(figsize=(10,6))
ax = Axes3D(figure)
ax.scatter(confirmed_cases['dl'], confirmed_cases['mh'], confirmed_cases['rj'])
plt.show()
```
